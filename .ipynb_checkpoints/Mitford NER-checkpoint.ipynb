```python
import os
import re
import pandas as pd
import spacy
from spacy import displacy
from segtok.segmenter import split_single
from flair.models import SequenceTagger
from flair.data import Sentence

tagger = SequenceTagger.load('ner')
nlp = spacy.load('en')
pd.set_option('chained_assignment', None)
```

```python
data_dir = 'mitford-letters'
clean_text_dir = '../mitford_letters_clean'
gold_standard_dir = '../mitford_letters_gs'
spacy_dir = '../spacy_mitford'
flair_dir = '../flair_mitford'
letters = os.listdir(clean_text_dir)
```

```python
for letter in letters[0:1]:
    with open(os.path.join(clean_text_dir, letter), 'r') as f:
        text = f.read()
        
    sentences = [Sentence(sent, use_tokenizer=True) for sent in split_single(text)]
    SequenceTagger = SequenceTagger.load('ner')
    tagger.predict(sentences)
    text = []
    tags = []
    begin_offsets = []
    end_offsets = []
    offset = 0
    for s in sentences:
        for e in s.get_spans('ner'):
            if e.tag in ['PER', 'LOC']:
                tags.append(e.tag)
                text.append(e.text)
                begin_offsets.append(e.start_pos + offset)
                end_offsets.append(e.end_pos + offset)
        offset += len(s.to_original_text())

    flair_df = pd.DataFrame({'tag': tags,
                            'begin_offsets': begin_offsets,
                            'end_offset': end_offsets,
                            'text': text})
    flair_df['tag'].loc[flair_df['tag'] == 'LOC'] = 'place'
    flair_df['tag'].loc[flair_df['tag'] == 'PER'] = 'person'
    flair_df.sort_values(by=['begin_offsets'], inplace=True)
    flair_df.to_csv(os.path.join(flair_dir, letter), index=False)
    flair_df
```

```python
for letter in letters:
    with open(os.path.join(clean_text_dir, letter), 'r') as f:
        text = f.read()
    
    text = text.strip()
    doc = nlp(text)
    ents = doc.ents
    ents = [(e.text, e.start_char, e.end_char, e.label_) for e in doc.ents if e.label_ in ['PERSON', 'GPE']]
    
    tags = []
    begin_offsets = []
    end_offsets = []
    text = []
    for e in ents:
        text.append(e[0])
        tags.append(e[-1])
        begin_offsets.append(e[1])
        end_offsets.append(e[2])
        
    spacy_df = pd.DataFrame({'tag': tags,
                            'begin_offsets': begin_offsets,
                            'end_offset': end_offsets,
                            'text': text})
    spacy_df['tag'].loc[spacy_df['tag'] == 'GPE'] = 'place'
    spacy_df['tag'].loc[spacy_df['tag'] == 'PERSON'] = 'person'
    spacy_df.to_csv(os.path.join(spacy_dir, letter), index=False)
```
